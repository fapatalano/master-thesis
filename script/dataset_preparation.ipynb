# Benchmark dataset

This dataset needs to satisfy some criteria:
- Structures need to be homodimer
- Xray crystallography
- Resolution <= 2.5
- No DNA
- No mutation
- Each chain has to be 50 long at least

After that structures are going to be filtered :
- Redundant structures were removed with a sequence similarity >=30%
- Sequence similarity between the chains of the homodimer need to be >= 95%

```python
import glob
import pandas as pd

path = '/home/fapatalano/Desktop/dataset/'
df = pd.concat(map(pd.read_csv,  glob.glob(f'{path}/*csv')),ignore_index=True)
df1 = df.iloc[::, 0:8].copy()
df1.drop(df1[df1['Entry ID'].isna()].index,inplace=True)
df1.drop(df1[df1['Oligomeric State']!='Homo 2-mer'].index,inplace=True)
df1.drop(df1[df1['Total Number of Polymer Instances (Chains) per Assembly']!= 2].index,inplace=True)
df1.drop(columns=['Entry ID'],inplace=True)
df1.reset_index(drop=True, inplace=True)
# df1.to_csv(path+'out.csv', index=False)
```

```python
f = open(path+"all_seq.fasta", "w")
for row in df1.itertuples():
    f.write(f'>{row[1]}\n{row[3]}\n')
f.close()
```

```python
from Bio import PDB
import Bio.PDB.Polypeptide as PP
import os

parser = PDB.PDBParser()
directory = '/home/fapatalano/Desktop/tesi/dataset/pdb_file/'
for filename in os.listdir(directory):
    sequence_1 =''
    sequence_2 =''
    if filename.endswith('.pdb'):
        pdb_code= filename[:4]
        structure = parser.get_structure(pdb_code, (os.path.join(directory, filename)))
        chains= PDB.Selection.unfold_entities(structure[0], 'C')
        residues_1= PDB.Selection.unfold_entities(chains[0], 'R')
        residues_2= PDB.Selection.unfold_entities(chains[1], 'R')
        for residue_1 in residues_1:
            s = PP.three_to_one(residue_1.resname)
            sequence_1+=s
        for residue_2 in residues_2:
            s2 = PP.three_to_one(residue_2.resname)
            sequence_2+=s2
        print(pdb_code)
        print(sequence_1)
        print(sequence_2)
```
